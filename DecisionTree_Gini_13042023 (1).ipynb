```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df=pd.read_csv('Gini.csv',index_col=0)
```

```python
df.head(10)
```

```python
df['Weather'].replace('Sunny',0,inplace=True)
```

```python
df['Weather'].replace('Windy',1,inplace=True)
df['Weather'].replace('Rainy',2,inplace=True)
```

```python
df.head(10)
```

```python
df['Parents'].replace('No',0,inplace=True)
df['Parents'].replace('Yes',1,inplace=True)
```

```python
df.head(10)
```

```python
df['Money'].replace('Poor',0,inplace=True)
df['Money'].replace('Rich',1,inplace=True)
```

```python
df.head(10)
```

```python
df['Decision'].replace('Cinema',0,inplace=True)
df['Decision'].replace('Tennis',1,inplace=True)
df['Decision'].replace('Stay-In',2,inplace=True)
df['Decision'].replace('Shopping',3,inplace=True)
```

```python
df.head(10)
```

```python
x=df.iloc[:,:3]
y=df.iloc[:,-1]
```

```python
y.head(10)
```

```python
from sklearn import tree
#display tree
import pydotplus
from io import StringIO
from IPython.display import display,Image
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
clf=tree.DecisionTreeClassifier(criterion='gini',random_state=0)
```

```python
clf=clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
print(y_pred,y_test)
```

```python
print("The accuracy of the Decision Tree classifier on training data is     {:.2f}".format(clf.score(x_train, y_train)))
print("The accuracy of the Decision Tree classifier on test data is     {:.2f}".format(clf.score(x_test, y_test)))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))  
print("\nClassiification Report:")
print(classification_report(y_test, y_pred))  
```
